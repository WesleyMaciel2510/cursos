#Descrição do Problema:

A série de Fibonacci é uma sequência de números onde cada número é a soma dos dois anteriores, começando com 0 e 1. Assim, os primeiros números da série são 0, 1, 1, 2, 3, 5, 8, 13, e assim por diante. Esta série tem aplicações em diversas áreas, incluindo matemática computacional, economia, biologia, entre outras.

O objetivo deste exercício é implementar uma função recursiva em Python que calcule o n-ésimo número da série de Fibonacci. A função deve ser definida de forma que receba um inteiro n como argumento e retorne o n-ésimo termo da série de Fibonacci.

#Requisitos do Exercício:

2.    A função deve ter uma condição base para os casos onde n é 0 ou 1, retornando 0 e 1, respectivamente.

3.   Para valores de n maiores que 1, a função deve chamar a si mesma para calcular os dois termos anteriores da série e retornar a soma desses dois termos.

4.   Inclua testes no seu código para verificar se a função está correta, chamando a função fibonacci(n) com diferentes valores de n e imprimindo os resultados.

#Resolvendo a série de Fibonacci com o uso de funções recursivas (com e sem memoization)

### Sem *memoization*

```python
# Solução sem memoization
def fibonacci(num):
  if (num <= 1):
    return num
  else:
    return fibonacci(num-1) + fibonacci(num-2)
```

### Com *memoization*

```python
# Solução com memoization
def fibonacci_memoization(num, memoization={}):
  if num in memoization:
    return memoization[num]
  if num <= 1:
    return num
  memoization[num] = fibonacci_memoization(num-1, memoization) + fibonacci_memoization(num-2, memoization)
  return memoization[num]
```

### Função para cálculo do tempo

```python
import time

def measure_time(func, n):
    start_time = time.time()
    result = func(n)
    end_time = time.time()
    return result, end_time - start_time
```

### Avaliação do tempo sem *memoization*

```python
valores = [10, 20, 30, 40]
resultado_sem = []

for valor in valores:
  result_no_memo, time_no_memo = measure_time(fibonacci, valor)
  resultado_sem.append(time_no_memo)
  print (f"result_no_memo, {format(time_no_memo, '.8f')}")
```

### Avaliação do tempo com *memoization*

```python
valores = [10, 20, 30, 40]
resultado_com = []

for valor in valores:
  result_no_memo, time_no_memo = measure_time(fibonacci_memoization, valor)
  resultado_com.append(time_no_memo)
  print (f"result_no_memo, {format(time_no_memo, '.8f')}")
```

###Análise dos Resultados:

A função fibonacci sem memoização exibe um aumento exponencial no tempo de execução conforme o tamanho da entrada cresce. Este comportamento é típico para implementações recursivas puras de algoritmos exponenciais, como o cálculo de Fibonacci.

Por outro lado, a função fibonacci_memoization com memoização mantém um tempo de execução extremamente baixo e praticamente constante, independentemente do tamanho da entrada. ###Conclusão:

A memoização provou ser uma técnica extremamente eficaz para otimizar funções recursivas com sobreposição substancial de subproblemas, como é o caso da sequência de Fibonacci. A redução no tempo de execução é notável, especialmente para entradas maiores, tornando a abordagem com memoização altamente recomendada para este tipo de tarefa computacional.

Esta análise demonstra claramente o poder da memoização para melhorar a eficiência de algoritmos recursivos, permitindo que operações potencialmente custosas sejam executadas em uma fração do tempo que seria necessário sem essa técnica.
