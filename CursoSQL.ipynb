#Criação das tabelas

Tabela Cliente:

```
CREATE TABLE Cliente (
    IDCliente INT PRIMARY KEY,
    Nome VARCHAR(100),
    NumeroComentarios INT,
    NumeroLikes INT,
    Localizacao VARCHAR(100),
    Genero VARCHAR(20)
);
```

````python
cliente_sql = "CREATE TABLE Cliente (IDCliente INT PRIMARY KEY, Nome VARCHAR(100), NumeroComentarios INT, NumeroLikes INT, Localizacao VARCHAR(100), Genero VARCHAR(20));"
````

````python
cursor.execute(cliente_sql)
````

Tabela Comentario:

```
CREATE TABLE Comentario (
    IDComentario INT PRIMARY KEY,
    TextoOriginal TEXT,
    TextoTraduzido TEXT,
    TituloOriginal VARCHAR(200),
    TituloTraduzido VARCHAR(200),
    IdiomaTexto VARCHAR(50),
    DataPublicacao DATE,
    NumeroLikes INT,
    Avaliacao INT,
    URL VARCHAR(100),
    IDCliente INT,
    FOREIGN KEY (IDCliente) REFERENCES Cliente(IDCliente),
);
```



````python
comentario_sql = "CREATE TABLE Comentario (IDComentario INT PRIMARY KEY, TextoOriginal TEXT, TextoTraduzido TEXT, TituloOriginal VARCHAR(200), TituloTraduzido VARCHAR(200), IdiomaTexto VARCHAR(50), DataPublicacao DATE, NumeroLikes INT, Avaliacao INT, URL VARCHAR(100), IDCliente INT, FOREIGN KEY (IDCliente) REFERENCES Cliente(IDCliente));"
````

````python
cursor.execute(comentario_sql)
````

````python
conexao.commit()
````

# Valores `NULL`

Os campos não preenchidos são representados pelo valor `NULL`.

Não é possível testar valores `NULL` com os operadores de comparação: `>`ou `=`.

É necessário usar os operadores `IS NULL` e `IS NOT NULL`.

````python
cliente11 = "INSERT INTO Cliente (IDCliente, Nome, NumeroComentarios, NumeroLikes, Localizacao, Genero) VALUES (11, 'Ana Silva', 15, 120, NULL, 'Feminino');"
````

````python
cursor.execute(cliente11)
````

````python
df_cliente = pd.read_sql_query(consulta, conexao)
````

````python
df_cliente
````

# Funções

O `SQL` possui um conjunto de funções que podem ser usadas durante as consultas ao banco de dados.

## `MAX` e `MIN`

````python
consulta = "SELECT Nome, MAX(NumeroComentarios) FROM Cliente;"
````

````python
cursor.execute(consulta)
````

````python
df_cliente = pd.read_sql_query(consulta, conexao)
df_cliente
````

## `COUNT`, `AVG` e `SUM`

````python
consulta = "SELECT COUNT(*) FROM Comentario;"
````

````python
df_comentario = pd.read_sql_query(consulta, conexao)
df_comentario
````

# Operador `Group By`

A instrução `Group By` agrupa as linhas que possuem os mesmos valores em linhas de resumo.

As instruções `Group By` são usualmente empregadas com funções agregadas (`MAX`, `MIN`, `COUNT`, `SUM`, `AVG`).

````python
consulta = "SELECT Genero, Count(*) FROM Cliente GROUP BY (Genero);"
````

````python
df_cliente = pd.read_sql_query(consulta, conexao)
df_cliente
````

# Operador de Junção - `JOIN`

Um `JOIN` é usado para combinar linhas de duas ou mais tabelas com base no relacionamento entre colunas.
