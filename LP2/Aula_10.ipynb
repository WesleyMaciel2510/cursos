#Ponteiros

Ponteiros são variáveis que armazenam **endereços de memória**, enquanto as variáveis comuns armazenam valores numéricos ou caracteres.

Exemplo de memória:

memoria.png

O endereço de memória corresponde, normalmente, ao endereço de memória de outra variável.

Esta é a razão pela qual costumamos afirmar que um ponteiro *aponta* para uma variável.

Um ponteiro pode apontar para uma variável de qualquer tipo, incluindo também estruturas de dados ou outros ponteiros.

Importância de ponteiros:

*   Algumas operações só podem ser executadas através de ponteiros. Por exemplo, ponteiros são necessários para a passagem de parâmetros por referência nas funções.
*   Ponteiros estão diretamente associados com a manipulação de vetores, matrizes e estruturas.
*   Os ponteiros permitem realizar alocação dinâmica de memória e manipular estruturas de dados dinâmicas.
*   A utilização de ponteiros produz um código compacto e eficiente. Apesar de que para programadores iniciantes, o uso de ponteiros pode gerar alguma confusão.

## Declaração de Ponteiros

A declaração de ponteiros é realizada com o operador **\*** e o operador **&** é utilizado para obter o endereço de memória de uma variável.

A sintaxe para a criação de um ponteiro é a seguinte:

```
<tipo-de-dados>   * <nome-ponteiro>;
```



Como é possível observar, um ponteiro sempre deve estar associado a um tipo de dados e só pode apontar para variáveis que sejam do seu tipo. 

Exemplo de declaração de ponteiros:



```
int *ptr;
```



Neste exemplo, o ponteiro ``ptr`` somente poderia apontar para variáveis do tipo ``int``.  

Porém, é possível criar *ponteiros genéricos*, ou seja, ponteiros que não possuem um tipo prévio de dados. 

Para isso, basta declarar o ponteiro como sendo do tipo ``void``. Assim, o ponteiro genérico pode apontar para qualquer tipo de variável.

Abaixo um exemplo de acesso a uma variável através do seu endereço de memória:

````python
%%writefile exemplo1-aula10.c
#include <stdio.h>

int main () {
  int *ptr;
	int valor = 100;

	ptr = &valor;

	printf("Endereço = %p \n", &valor);
	printf("Endereço = %p \n", ptr);
	printf("Valor = %d", *ptr);

	return 0;
}
````

````python
%%shell
gcc exemplo1-aula10.c -o exemplo1-aula10
./exemplo1-aula10
````

A variável `valor` foi inicializada com o valor 100 e o ponteiro `ptr` apontou para o endereço de `valor`.

Deve-se observar que os ponteiros somente aceitam endereços de memória.

## Ponteiros para Ponteiros

É permitido que um ponteiro aponte para outro ponteiro. 

Quando isto ocorre, nós dizemos que criamos ponteiro para ponteiro.

Um ponteiro para ponteiro é declarado com dois asteriscos.

Sintaxe de criação de ponteiro para ponteiro:



```
<tipo-de-dados> **pptr;
```



O código abaixo mostra um exemplo de uso de ponteiros para ponteiro.

````python
%%writefile exemplo2-aula10.c
#include <stdio.h>
#include <stdlib.h>

int main () {
	int * ptr;
	int ** pptr;

	int valor = 100;

	ptr = &valor;
	pptr = &ptr;

	printf("Endereço de ptr = %p \n", &ptr);
	printf("Endereço de pptr = %p \n", &pptr);

	printf("Valor ptr = %d \n", *ptr);
	printf("Valor pptr = %d \n", **pptr);

	return 0;
}
````

````python
%%shell
gcc exemplo2-aula10.c -o exemplo2-aula10
./exemplo2-aula10
````
