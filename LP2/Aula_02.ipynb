#Fundamentos da Linguagem C

##Linguagem C



*   Um programa em C consiste de um ou mais blocos chamados de **funções**.
*   Um programa em C deve conter uma função chamada **main**.
*   A função **main** indica o ponto de início da execuçãod o programa.



## Estrutura de um programa em C


```
#include <stdio.h>

definição de constantes

lista de blocos de funções

int main() {
  declaração de variáveis
  comandos
}
```



## Comandos

Os comandos simples são apresentados em uma única linha e terminados com um ponto-e-vírgula.

```
printf ("Vamos iniciar...");
```

Os blocos são grupos de comandos, onde as chaves são usadas para indicar o início e o término do bloco.



```
{
  i = 3;
  printf("%d\n", i);
  i = i + 1;
}
```



# Variáveis

Variável é um espaço na memória onde iremos armazenar um valor e que será usado no programa.

As variáveis podem ser usadas para armazenar diferentes **tipos** de valores.

Cada variável usa uma determinada quantidade de memória para representar os seus valores.

Variáveis do mesmo tipo utilizam a mesma quantidade de *bytes*.

### Regras para declaração de variável



*   Qualquer sequência de letras, digitos, e \_, mas **deve começar** com uma letra ou com '_'.
*   Não são permitidos nomes ou palavras reservadas da linguagem.

Sugestões para nomes de variáveis:



*   Dê nomes significativos as variáveis (mas não muito longos);
*   Use nomes de variáveis do tipo i, j, k somente para variáveis tipo contadores; 
*   Pode-se usar letras maiúsculas ou '_' para juntar palavras. Exemplo: `valorInicial` ou `valor_inicial`.



# Constantes

Constantes são usadas para armazenar valores que **NÃO** podem ser modificados durante a execução de um programa.

Sintaxe:

`#define <nome_da_constante> <valor>`

Exemplos:



```
#define PI 3.14
#define NOTA 100
```

## Saída de Dados

O comando `printf()` é utilizado para imprimir mensagens e valores em uma variedade de formatos.

Exemplo para imprimir *string* (texto):

`printf ("Vamos iniciar...");`

É possível ainda usar comandos especiais:

`printf ("Linha 1\n");`

Se quisermos imprimir o valor de variáveis, será necessário incluir argumentos.


A primeira parte indica o tipo da variável e a segunda parte o nome da variável.

O tipo da variável é indicado por uma máscara.

Principais formatos:

*   %c: caracter.
*   %d: inteiro.
*   %f: real.


Exemplo [[1](https://www.inf.ufpr.br/cursos/ci067/Docs/NotasAula/notas-1_Programas_C.html)]:

```
#include $<$stdio.h$>$

int main() { 

int pera = 3; 

char qualidade = 'A'; 

float peso = 2.5; 


printf(``Existem %d peras de qualidade %c ``, pera, qualidade); 

printf(``pesando %f quilos.\n'', peso); 

printf(``O preco por quilo e %f, total e %f\n'', PRECO, peso * PRECO); 

}
```



## Entrada de Dados

O comando `scanf()` é usado para ler valores digitados no teclado.

São usadas as mesmas máscaras do comando de saída (`printf`).

`scanf(``%d'', &entrada);`

#Exemplo:

````python
%%writefile exemplo1-aula2.c

#include <stdio.h>

int main() {
    int ent;
    scanf("%d", &ent);
    printf ("Você tem %d anos.", 2021-ent);
    return 0;
}
````
